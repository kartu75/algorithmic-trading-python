```python
symbol ='AAPL'
ticker=yf.Ticker('AAPL')
one_yr_change=ticker.info.get('52WeekChange','N/A')
one_yr_change
```

```python
my_coloumn=['Ticker','Price','One Year Returns','Number of Shares to buy']
final_dataframe=pd.DataFrame(columns=my_coloumn)
for symbol in stocks['Ticker']:
  ticker=yf.Ticker(symbol)
  price = ticker.info.get('currentPrice') or ticker.info.get('regularMarketPrice', 'N/A')
  one_yr_change=ticker.info.get('52WeekChange','N/A')
  newrow={
    'Ticker':symbol,
    'Price':price,
    'One Year Returns':one_yr_change,
    'Number of Shares to buy': 'N/A'

  }
  final_dataframe = pd.concat([final_dataframe, pd.DataFrame([newrow])], ignore_index=True)
final_dataframe
```

```python
final_dataframe = final_dataframe[final_dataframe['One Year Returns'] != 'N/A'].reset_index(drop=True)
final_dataframe.sort_values('One Year Returns', ascending=False, inplace=True)
final_dataframe=final_dataframe[:50]
final_dataframe
```

```python
for i in range(len(final_dataframe)):
    final_dataframe.loc[i, 'Number of Shares to buy'] = math.floor(position_size / final_dataframe.loc[i, 'Price'])
final_dataframe    
```
